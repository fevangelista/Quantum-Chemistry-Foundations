## Fundamental Integrals

In this tutorial, we will derive various integrals of general 3D Gaussian functions (with positive exponents $\alpha_\mu > 0$):

\begin{equation}
\chi_\mu(\mathbf{r}) = N_{\alpha_\mu} e^{-\alpha_\mu (x^2 + y^2 + z^2)} = N_{\alpha_\mu} e^{-\alpha_\mu r^2}.
\end{equation}

Here we report some fundamental integrals of a 1D Gaussian over the interval $x \in [0,\infty)$ that will be useful in the derivations:

\begin{align}
I_1(\alpha) &= \int_0^\infty x e^{-\alpha x^2} dx = \frac{1}{2 \alpha},
\\
I_2(\alpha) &= \int_0^\infty x^2 e^{-\alpha x^2} dx = \frac{1}{4\alpha} \sqrt{\frac{\pi}{\alpha}},
\\
I_4(\alpha) &= \int_0^\infty x^4 e^{-\alpha x^2} dx = \frac{3}{8\alpha^2} \sqrt{\frac{\pi}{\alpha}}.
\end{align}

All these integrals are valid if **the parameter $\alpha$ is positive**. Also note that the integration limit in this case extends from 0 to infinity.

You can find more details about integrals of Gaussian functions on [this page](https://en.wikipedia.org/wiki/Gaussian_integral).

## Overlap Integral and Normalization Condition

The simplet integral, the overlap integral $\langle \chi_\mu | \chi_\nu \rangle$, is defined as:
\begin{align}
\langle \chi_\mu | \chi_\nu \rangle = \int \chi^*_\mu(\mathbf{r}) \chi_\nu(\mathbf{r}) d\mathbf{r} = N^*_{\alpha_\mu} N_{\alpha_\nu}
\int e^{-(\alpha_\mu + \alpha_\nu) r^2} d\mathbf{r}.
\end{align}

Since the integrand depends only on the radial distance $r$, this integral can be simplified by converting it to [spherical coordinates](https://en.wikipedia.org/wiki/Spherical_coordinate_system) via the change of coordinates $(x,y,z) \rightarrow (r,\theta,\phi)$.
Since $\theta$ and $\phi$ don't explicitly appear in the integral, we can effectively make the replacement:
\begin{equation}
\int f(x,y,z) d\mathbf{r} \rightarrow 4\pi \int_0^\infty f(r) dr,
\end{equation}
where the $4\pi$ factor comes from integrating over $\theta$ and $\phi$.
Written in spherical coordinates, the overlap integral reads:

\begin{align}
\langle \chi_\mu | \chi_\nu \rangle  & = N^*_{\alpha_\mu} N_{\alpha_\nu}
4\pi \underbrace{\int_0^\infty r^2  e^{-(\alpha_\mu + \alpha_\nu) r^2}dr }_{I_2(\alpha_\mu + \alpha_\nu) } \\
& = N^*_{\alpha_\mu} N_{\alpha_\nu}
\left(\frac{\pi}{\alpha_\mu + \alpha_\nu}\right)^{3/2}.
\end{align}

Here, we recognize that the integral over the radial distance is equal to the fundamental integral $I_2$ evaluated with an argument equal to the sum of the exponents.


From the normalization condition, $\langle \chi_\mu | \chi_\mu \rangle = 1$ (which we decided to impose), it follows that:

\begin{align}
\langle \chi_\mu | \chi_\mu \rangle  & =  1 = |N_{\alpha_\mu}|^2 \left(\frac{\pi}{2 \alpha_\mu}\right)^{3/2}
\Rightarrow |N_{\alpha_\mu}|^2 = \left(\frac{2 \alpha_\mu}{\pi}\right)^{3/2},
\end{align}

and we can use this result to derive an expression for the normalization constant (assuming a phase factor of +1):
\begin{equation}
N_{\alpha_\mu} = 
\left(\frac{2\alpha_\mu}{\pi}\right)^{3/4}.
\end{equation}

From this result, we get our final expression for the integral:
\begin{equation}
\langle \chi_\mu | \chi_\nu \rangle  = (\alpha_\mu \alpha_\nu)^{3/4} \left(\frac{2}{\alpha_\mu + \alpha_\nu}\right)^{3/2}.
\end{equation}

## Potential energy integrals

We will proceed to evaluate the potential energy integrals next since they are just a bit more complex than the overlap integrals.
We start by writing the integral explicitly:

\begin{align}
\langle \chi_\mu | \hat{V} | \chi_\nu \rangle  & = -\int \chi^*_\mu(\mathbf{r}) \frac{1}{r} \chi_\nu(\mathbf{r}) d\mathbf{r} = -N^*_{\alpha_\mu} N_{\alpha_\nu}
\int \frac{1}{r} e^{-(\alpha_\mu + \alpha_\nu) r^2} d\mathbf{r} .
\end{align}

In the next step, we convert the integral to spherical coordinates, recognizing that this integral can be related to the fundamental integral $I_1$:
\begin{align}
\langle \chi_\mu | \hat{V} | \chi_\nu \rangle
& = -N^*_{\alpha_\mu} N_{\alpha_\nu}
4\pi \underbrace{\int_0^\infty r  e^{-(\alpha_\mu + \alpha_\nu) r^2} dr}_{I_1(\alpha_\mu + \alpha_\nu)}  \\
& = -N^*_{\alpha_\mu} N_{\alpha_\nu}  
\frac{2\pi}{\alpha_\mu + \alpha_\nu} .
\end{align}

Inserting the definition of the normalization constants we arrive at the final expression for the potential energy integrals:

\begin{align}
\langle \chi_\mu | \hat{V} | \chi_\nu \rangle
& = -4 \sqrt{\frac{2}{\pi}}  \frac{(\alpha_\mu \alpha_\nu)^{3/4} }{\alpha_\mu + \alpha_\nu} .
\end{align}

## Kinetic energy integrals

We can proceed similarly to evaluate the kinetic energy integrals

\begin{equation}
\langle \chi_\mu | \hat{T} | \chi_\nu \rangle   = -\frac{1}{2} \int \chi^*_\mu(\mathbf{r}) \nabla^2 \chi_\nu(\mathbf{r}) d\mathbf{r} .
\end{equation}

Focusing on the integral, we first want to convert the integral and the expression for $\nabla^2$ to spherical coordinates. When the integrand does not depend on $\theta$ and $\phi$, we can use the simplified form of the Laplacian:

\begin{equation}
\nabla^2 f(r) = \frac{1}{r^2} \frac{\partial}{\partial r} \left(r^2 \frac{\partial f(r)}{\partial r} \right) .
\end{equation}

Converting this integral to spherical coordinates, we get

\begin{align}
\int \chi^*_\mu(\mathbf{r}) \nabla^2 \chi_\nu(\mathbf{r}) d\mathbf{r}  =  4 \pi N^*_{\alpha_\mu} N_{\alpha_\nu}
\int_0^\infty  r^2 e^{-\alpha_\mu r^2} \frac{1}{r^2} \frac{\partial}{\partial r} \left(r^2 \frac{\partial}{\partial r} e^{-\alpha_\nu r^2} \right) dr .
\end{align}

Let's evaluate the Laplacian explicitly:

\begin{equation}
\frac{1}{r^2} \frac{\partial}{\partial r} \left(r^2 \frac{\partial}{\partial r} e^{-\alpha_\nu r^2} \right)
= \frac{1}{r^2} \frac{\partial}{\partial r} \left(-2 \alpha_\nu r^3 e^{-\alpha_\nu r^2} \right)
= (-6 \alpha_\nu  + 4 \alpha_\nu^2 r^2) e^{-\alpha_\nu r^2}  .
\end{equation}

Using this result, we can write the kinetic energy integral as:

\begin{align}
\langle \chi_\mu | \hat{T} | \chi_\nu \rangle  & =  -\frac{4\pi}{2} N^*_{\alpha_\mu} N_{\alpha_\nu}
\int_0^\infty e^{-\alpha_\mu r^2} (-6 \alpha_\nu r^2  + 4 \alpha_\nu^2 r^4) e^{-\alpha_\nu r^2}  dr \\
& =  -2 \pi N^*_{\alpha_\mu} N_{\alpha_\nu}  \left[
-6 \alpha_\nu I_2(\alpha_\mu + \alpha_\nu)
+ 4 \alpha_\nu^2 I_4(\alpha_\mu + \alpha_\nu)
\right] \\
& = 
12 \pi N^*_{\alpha_\mu} N_{\alpha_\nu} \alpha_\nu I_2(\alpha_\mu + \alpha_\nu)
- 8 \pi N^*_{\alpha_\mu} N_{\alpha_\nu} \alpha_\nu^2 I_4(\alpha_\mu + \alpha_\nu)
\end{align}

The first term can be simplified to:
\begin{align}
12 \pi N^*_{\alpha_\mu} N_{\alpha_\nu} \alpha_\nu I_2(\alpha_\mu + \alpha_\nu)
& = 12 \pi
\left(\frac{2\alpha_\mu}{\pi}\right)^{3/4} \left(\frac{2\alpha_\nu}{\pi}\right)^{3/4}  \alpha_\nu \frac{1}{4 (\alpha_\mu + \alpha_\nu)} \sqrt{\frac{\pi}{\alpha_\mu + \alpha_\nu}} \\
& =  \left(4 \alpha_\mu \alpha_\nu\right)^{3/4} \frac{3 \alpha_\nu}{(\alpha_\mu + \alpha_\nu)^{3/2}},
\end{align}

and similarly for the second term:

\begin{align}
- 8 \pi N^*_{\alpha_\mu} N_{\alpha_\nu} \alpha_\nu^2 I_4(\alpha_\mu + \alpha_\nu)
& =
- 8 \pi \left(\frac{2\alpha_\mu}{\pi}\right)^{3/4} \left(\frac{2\alpha_\nu}{\pi}\right)^{3/4}  \alpha_\nu^2 \frac{3}{8(\alpha_\mu + \alpha_\nu)^2} \sqrt{\frac{\pi}{\alpha_\mu + \alpha_\nu}} \\
& = - \left(4 \alpha_\mu \alpha_\nu \right)^{3/4}  \frac{3 \alpha_\nu^2 }{(\alpha_\mu + \alpha_\nu)^{5/2}} .
\end{align}

Putting these two results together, we arrive (after a few complicated steps) to our final expression for the kinetic energy integral:

\begin{align}
\langle \chi_\mu | \hat{T} | \chi_\nu \rangle  & =  \left(4 \alpha_\mu \alpha_\nu\right)^{3/4} \frac{3 \alpha_\nu}{(\alpha_\mu + \alpha_\nu)^{3/2}} - \left(4 \alpha_\mu \alpha_\nu \right)^{3/4}  \frac{3 \alpha_\nu^2 }{(\alpha_\mu + \alpha_\nu)^{5/2}} \\
& =  3 \left(4 \alpha_\mu \alpha_\nu\right)^{3/4} \left[
    \frac{\alpha_\nu}{(\alpha_\mu + \alpha_\nu)^{3/2}} - \frac{\alpha_\nu^2 }{(\alpha_\mu + \alpha_\nu)^{5/2}}
\right] \\
& =  3 \times 2 \sqrt{2} \left(\alpha_\mu \alpha_\nu\right)^{3/4} \left[
    \frac{\alpha_\nu(\alpha_\mu + \alpha_\nu)}{(\alpha_\mu + \alpha_\nu)^{5/2}} - \frac{\alpha_\nu^2 }{(\alpha_\mu + \alpha_\nu)^{5/2}}
\right] \\
& =  6 \sqrt{2} \left(\alpha_\mu \alpha_\nu\right)^{3/4} 
    \frac{\alpha_\nu\alpha_\mu}{(\alpha_\mu + \alpha_\nu)^{5/2}}\\
& =  6 \sqrt{2}
    \frac{\left(\alpha_\mu \alpha_\nu\right)^{7/4}}{(\alpha_\mu + \alpha_\nu)^{5/2}} .
\end{align}
